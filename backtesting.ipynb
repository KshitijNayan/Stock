```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your backtest results CSV
backtest_data = pd.read_csv("backtest_results.csv")

# Clean column names in case of extra spaces
backtest_data.columns = backtest_data.columns.str.strip()

# Extract relevant data
months = backtest_data["Month"]
returns = backtest_data["Returns"]
sharpe_ratios = backtest_data["Sharpe Ratio"]

# Function to add data labels
def add_labels(x, y):
    for i in range(len(x)):
        plt.text(i, y[i] + 0.1, f"{y[i]:.2f}", ha='center', fontsize=8)

# Plot Monthly Returns with average line and labels
plt.figure(figsize=(12, 5))
plt.bar(months, returns, color='skyblue')
avg_return = returns.mean()
plt.axhline(avg_return, color='orange', linestyle='--', label=f"Avg Return: {avg_return:.2f}%")
add_labels(months, returns)
plt.xlabel("Month")
plt.ylabel("Returns (%)")
plt.title("Monthly Returns Over 2 Years")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

# Plot Sharpe Ratio with average line and labels
plt.figure(figsize=(12, 5))
plt.bar(months, sharpe_ratios, color='salmon')
avg_sharpe = sharpe_ratios.mean()
plt.axhline(avg_sharpe, color='orange', linestyle='--', label=f"Avg Sharpe: {avg_sharpe:.2f}")
add_labels(months, sharpe_ratios)
plt.xlabel("Month")
plt.ylabel("Sharpe Ratio")
plt.title("Monthly Sharpe Ratios Over 2 Years")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
